### 1-1) Linear Regression from scratch

```python
import numpy as np
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
x = np.random.randn(1,100)
y = 3 * x + np.random.rand(1,100) * 5
```

```python
plt.scatter(x,y)
```

```python
print (x.shape)
print (y.shape)
```

```python
def hypothesis(x, w, b):
    pred = np.dot(x, w) + b
    return pred
```

```python
def cost(x, w, b, y):
    loss = 1/2*(hypothesis(x, w, b) - y)**2
    cost = 1/100*np.sum(loss, axis=1, keepdims=True)
    return cost
```

```python
def derivative(x, w, b, y):
    dw = (1/100*np.dot(x, (hypothesis(x, w, b) - y).T))[0][0]
    db = 1/100*np.sum(hypothesis(x, w, b) - y)
    return dw, db
```

```python
def update(x, w, b, y, alpha):
    w = w - alpha*(derivative(x, w, b, y)[0])
    b = b - alpha*(derivative(x, w, b, y)[1])
    return w, b
```

```python
w = 0
b = 0
```

```python
for i in range(1000):
    print ('cost =',cost(x, w, b, y)[0][0])
    w, b = update(x, w, b, y, alpha=0.01)
    print ('w =',w)
    print ('b =',b)
    print ('\n')
```

```python
plt.scatter(x,y)

line_x = np.linspace(-4,4,100)
plt.plot(line_x, line_x * w + b)
plt.show()

print ('w =',w)
print ('b =',b)
```

### 1-2) Linear Regression with scikit-learn

```python
from sklearn.linear_model import LinearRegression
```

```python
reshaped_x = x.reshape((100,1))
reshaped_y = y.reshape((100,1))
model = LinearRegression()
model.fit(reshaped_x,reshaped_y)
```

```python
print ('w =', model.coef_)
print ('b =', model.intercept_)
```
